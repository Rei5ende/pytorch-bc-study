# Pytorch 기초

```python
import torch

# 텐서 만들기
a = torch.tensor([1.0, 2.0, 3.0])
b = torch.tensor([4.0, 5.0, 6.0])

print("a:", a)
print("b:", b)
print("a + b:", a + b) #tensor([5., 7., 9.]) ✅ 원소별 덧셈
print("a * b:", a * b) #tensor([4., 10., 18.])   ✅ 원소별 곱셈


# 텐서 shape 확인
x = torch.zeros(3, 4) #✅ 3행 4열 영텐서
                    # 영텐서 — 모든 원소가 0
#torch.ones(3, 4)    # 일텐서 — 모든 원소가 1
#torch.rand(3, 4)    # 랜덤텐서 — 0~1 사이 랜덤값
#torch.eye(3)        # 단위행렬 — 대각선만 1, 나머지 0
print("shape:", x.shape)
print(x)
```

#Autograd (자동미분)

```python
import torch

# requires_grad=True → 이 텐서의 기울기를 추적하겠다
x = torch.tensor(3.0, requires_grad=True)

# y = x^2
y = x ** 2

# 역전파 (기울기 계산)
y.backward()

# dy/dx = 2x = 2*3 = 6
print("x:", x)
print("y:", y)
print("dy/dx:", x.grad)
```
